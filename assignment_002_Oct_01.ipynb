# 6. Find the answer as well as find out the reason behind the result? -

Heap memory is the dynamic memory allocation that uses a garbage collection algorithm (called Garbage Collector) that keeps the Heap memory clean and removes objects that are not needed anymore. Most of the data objects I enter are stored in this memory.

To optimize memory allocation, Python does a process called “interning.” For some objects, Python only stores one object on Heap memory and ask different variables to point to this memory address if they use those objects. The objects that Python does interning on them are integer numbers (between -5 and 256], boolean, and some strings. Interning does not apply to other types of objects such as large integers, most strings, floats, lists, dictionaries, tuples.
